```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('raw_nino_flats_dataset.csv')
df.head(2)
```

### Избавляемся от лишних букв

```python
df.isna().sum()
```

Где нет NaN значений можем перевести строки в числа используя метод `replace()`. Для остальных используем цикл, где выполняем действия для непустых значений

```python
df['roomsCount'] = [int(item.replace('-комн', '')) for x, item in df['roomsCount'].iteritems()]
df.head(2)
```

```python
df['price'] = [int(item.replace(' ₽/мес.', '').replace(' ', '')) for x, item in df['price'].iteritems()]
df.head(2)
```

```python
deposits = []
for x, item in df['deposit'].iteritems():
    if pd.notna(item):
        deposits.append(float(item.replace('Без залога', '0').replace('₽', '').replace('Залог', '').replace(' ', '')))
    else:
        deposits.append(item)
df['deposit'] = deposits
df.head(2)
```

```python
comissions = []
for x, item in df['comission'].iteritems():
    if pd.notna(item):
        comissions.append(float(item.replace('без комиссии', '0').replace('%', '').replace('комиссия ', '')))
    else:
        comissions.append(item)
df['comission'] = comissions
df.head(2)
```

```python
prepays = []
for x, item in df['prepay'].iteritems():
    if pd.notna(item):
        prepays.append(int(item.replace('без предоплаты', '0').replace('предоплата за ', '').replace(' месяца', '').replace(' месяц', '')))
    else:
        prepays.append(item)
df['prepay'] = prepays
df.head(2)
```

```python
total_areas = []
for x, item in df['total_area'].iteritems():
    if pd.notna(item):
        total_areas.append(float(item.replace(' м²', '').replace(',', '.')))
    else:
        total_areas.append(item)
df['total_area'] = total_areas
df.head(2)
```

```python
living_areas = []
for x, item in df['living_area'].iteritems():
    if pd.notna(item):
        living_areas.append(float(item.replace(' м²', '').replace(',', '.')))
    else:
        living_areas.append(item)
df['living_area'] = living_areas
df.head(2)
```

```python
kitchen_areas = []
for x, item in df['kitchen_area'].iteritems():
    if not pd.isna(item):
        kitchen_areas.append(float(item.replace(' м²', '').replace(',', '.')))
    else:
        kitchen_areas.append(item)
df['kitchen_area'] = kitchen_areas
df.head(2)
```

Этажи разделяем на два столбца: `floors` - номер этажа квартиры, и `floors_cnt` - общее количество этажей в доме.

```python
floors = []
floor_cnts = []
for x, item in df['floor'].iteritems():
    if not pd.isna(item):
        floor, floor_cnt = item.split(' из ')
        floors.append(int(floor))
        floor_cnts.append(int(floor_cnt))
    else:
        floors.append(item)
        floor_cnts.append(item)
df['floor'] = floors
df['floor_cnt'] = floor_cnts
df.head(2)
```

Время до метро также разделяем на два слобца: `und_time` - время до метро, `und_type` - тип транспорта до метро. К тому же здесь можно сформировать новый столбец `und_dist` - примерное расстояние до метро. При `und_type = 'пешком'` считаем скорость **90** м/мин, при  `und_type = 'на транспорте'` считаем скорость **700** м/мин.

```python
und_times = []
und_types = []
und_dists = []
for x, item in df['undergroundTime'].iteritems():
    if not pd.isna(item):
        und_time, und_type = item.split(' мин. ')
        und_times.append(int(und_time.replace('<', '')))
        und_types.append(und_type)
        if und_type == 'пешком':
            und_dists.append(int(und_time.replace('<', '')) * 90)
        elif und_type == 'на транспорте':
            und_dists.append(int(und_time.replace('<', '')) * 700)
    else:
        und_times.append(item)
        und_types.append(item)
        und_dists.append(item)
df['und_time'] = und_times
df['und_type'] = und_types
df['und_dist'] = und_dists
df.head(2)
```

### Заполняем пропуски

```python
df.isna().sum()
```

```python
df.deposit.value_counts()
```

Видим, что у `deposit` больше всего значений 0, поэтому можно смело заполнить пропуски нулями

```python
df.deposit.fillna(0, inplace=True)
```

```python
df.comission.value_counts()
```

У `comission` значение 0 второе по популярности, но если на сайте ничего не было заполнено, будем расценивать это как ноль

```python
df.comission.fillna(0, inplace=True)
```

```python
df.prepay.value_counts()
```

У `prepay` практически все значения 1. Думаю, тут даже и комменитровать ничего не нужно

```python
df.prepay.fillna(1, inplace=True)
df.prepay.value_counts()
```

Для заполнения пропусков в `living_area` и `kitchen_area` формируем списки из `total_area` для каждой строки с пропуском умноженное на среднее отношение площади комнат и общей площади.

```python
liv_fill = round(df.total_area[df.living_area.isna()] * df.living_area.mean() / df.total_area.mean(), 1)
df.living_area.fillna(liv_fill, inplace=True)
```

```python
kitch_fill = round(df.total_area[df.kitchen_area.isna()] * df.kitchen_area.mean() / df.total_area.mean(), 1)
df.kitchen_area.fillna(kitch_fill, inplace=True)
```

Добавляем новые колонки: `first_pay` - сколько нужно будет заплатить сразу, сумма предоплаты, комиссии и депозита, `price_on_m` - цена за метр квадртаный.

```python
df['first_pay'] = df.price * (df.prepay + df.comission/100) + df.deposit
df['price_on_m'] = round(df.price / df.total_area, 2)
df.head(2)
```

```python
df.isna().sum()
```

Информацию о метро и годе постройки заполнять не будем, так как там довольно много пропущено, а заполнение средними значениями приведет к неточности, да и вообще заполнение чем-либо не имеет смысла.

### Оформляем датасет и загружаем в файл

Переименуем колонки и расставим их в более удобном порядке

```python
df = df.rename(columns={'roomsCount' : 'rooms', 'underground' : 'und'})
cols = ['rooms', 'price', 'price_on_m', 'first_pay', 'total_area', 'living_area', 'kitchen_area',
        'deposit', 'comission', 'prepay', 'und', 'und_time', 'und_type', 'und_dist', 'floor',
        'floor_cnt', 'build_year', 'lng', 'lat', 'link', 'tel', 'description']
df = df[cols]
df.head(2)
```

```python
df.to_csv('nino_flats_dataset.csv', index=False)
```
